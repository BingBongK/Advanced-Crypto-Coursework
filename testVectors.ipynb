```python
import json
import random
import galois 

tests = []
GF = galois.GF(2**8)


for _ in range(10):

    a = GF.Random()
    while a == 0:
        a = GF.Random()
    
    b = GF.Random()
    
    state = [random.randint(0,255) for _ in range(16)]
    key   = [random.randint(0,255) for _ in range(16)]

    expected = [int(a * GF((state[i]) ^ GF(key[i]))) for i in range(16)]

    tests.append({
        "a": int(a),
        "b": int(b),
        "state": state,
        "key": key,
        "expected": expected
    })

with open("test_vectors.txt", "w") as f:
    json.dump(tests,f,indent=2)
```
